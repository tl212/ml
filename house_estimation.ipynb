```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv("houselistings_simulated.csv")

df.head()
```

```python
for dt in ['SalesDate', 'ListingDate']:
    df[dt] = pd.to_datetime(df[dt])
```

```python
df['timetosell'] = (df['SalesDate'] - df['ListingDate']).dt.days
```

```python
xcolumns = df.columns[2: -1]
```

```python
xcolumns
```

```python
y = df['timetosell']
X = df[xcolumns]
```

```python
# to fit the dates into the algorithm find the day of the year and because the time is continuous we 
# need a circular representation. 
#Therefore the sin and cosine of the circle are used to represent the seasons

days = df['ListingDate'].dt.dayofyear * 360 /365

X['yearcos'] = np.cos(days)
X['yearsin'] = np.sin(days)
```
